```python
from scipy.stats import skewnorm
from sklearn.preprocessing import PowerTransformer
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#skewnorm.rvs es una función en el módulo scipy.stats de Python, que genera muestras aleatorias a partir de una distribución normal sesgada.
#En este caso se generaron 1000 muestras sesgadas con tamaño 1000
np.random.seed(12)
x = skewnorm.rvs(1000, size=1000)
```

```python
x
```

```python
# Histograma de la distribucion generada con 100 numeros randomicos en sesgo
sns.histplot(x)
plt.show()
```

##***Actividad Distribucion Uniforme***
----
> **1.** Hacer la distribución más normal usando raíz cuadrada.
* Dibujar la Distribución


```python
# Se genera la misma de distribucion de aleatorios con sesgo y se utiliza sqr para realizar la distribucion normal con raiz
dist_raiz = np.sqrt(x)
```

```python
dist_raiz
```

```python
# Se genera el histograma correspondiente de la distribucion de raiz
sns.histplot(dist_raiz)
```

##***Power Transformer YEO***
----
>**2.** Hacer la distribución más normal usando Power Transformer de sklearn y el método yeo-johnson

            a. Dibujar la nueva Distribución
            b. A cuánto equivale el punto 2.1 en la nueva transformación
            yeo : 0.9215448360095685
            c. Calcular manualmente el punto b y corroborar el resultado de Sklearn

```python
pt = PowerTransformer(standardize= False)
```

```python
pt.fit(x.reshape(-1,1))
```

```python
pt_trans = pt.transform(x.reshape(-1,1))
```

```python
# Se genera distribucion de power transformer
pt_trans
```

```python
# a Histograma de la nueva distribucion
sns.histplot(pt_trans)
```

##***Calculo manual metodo YEO***
----

```python
point = pt.transform([[2.1]])
```

```python
# b. A cuánto equivale el punto 2.1 en la nueva transformación
# yeo : 0.9215448360095685
point
```

```python
lambdas = pt.lambdas_
```

```python
lambdas
```

```python
point_yeo = ((((2.1 + 1)**lambdas) - 1) / lambdas)
```

```python
point_yeo
```

```python
# c. Calcular manualmente el punto b y corroborar el resultado de Sklearn
yeo = ((((2.1+1)**-0.37597537)-1)/-0.37597537)
yeo
```

##***Power Transformer Box-Cox***
----
>**2.** Hacer la distribución más normal usando Power Transformer de sklearn y el método box-cox

            a. Dibujar la nueva Distribución
            b. A cuánto equivale el punto 2.1 en la nueva transformación
            box-cox : 0.87674883
            c. Calcular manualmente el punto b y corroborar el resultado de Sklearn

```python
# Metodo perform a box-cox
pt_box = PowerTransformer(method='box-cox', standardize= False)
```

```python
pt_box.fit(x.reshape(-1,1))
```

```python
pt_box_trans = pt_box.transform(x.reshape(-1,1))
```

```python
pt_box_trans
```

```python
# a Histograma de la nueva distribucion con box-cox
sns.histplot(pt_box_trans)
```

##***Calculo manual metodos BOX-COT***
----

```python
# b. A cuánto equivale el punto 2.1 en la nueva transformación
# box-cox : 0.87674883
pt_box.transform([[2.1]])
```

```python
lambdas_box = pt_box.lambdas_
```

```python
lambdas_box
```

```python
box = (((2.1)**lambdas_box)-1)/(lambdas_box)
```

```python
# c. Calcular manualmente el punto b y corroborar el resultado de Sklearn
box
```
