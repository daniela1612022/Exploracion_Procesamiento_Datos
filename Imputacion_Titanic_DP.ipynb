```python
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('/content/titanic_1.csv')
```

```python
df
```

```python
im = imputer = SimpleImputer(strategy="mean")
```

```python
df['Age'] = im.fit_transform(df[['Age']]).ravel()
```

```python
imputer.statistics_
```

```python
df
```

```python
im2 = imputer = SimpleImputer(strategy="median")
```

```python
df['Age'] = im2.fit_transform(df[['Age']]).ravel()
```

```python
imputer.statistics_
```

```python
df
```

```python
im3 = imputer = SimpleImputer(strategy='most_frequent')
```

```python
df['Embarked'] = im3.fit_transform(df[['Embarked']]).ravel()
```
